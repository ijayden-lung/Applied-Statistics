```r
#Extract the data corresponding to species virginica to a separate data frame. Plot the numerical
#variables for this set in a matrix of plots.
iris.vir <- subset(iris, Species == 'virginica') 
str(iris.vir)
```

```r
plot(iris.vir[,1:4])
```

```r
scatterplot(Sepal.Length ~ Petal.Length, data = iris.vir)
```

```r
modelA <- lm(Sepal.Length ~ Petal.Length, data = iris.vir) 
summary(modelA)
```

```r
#Find the R2 and verify that for simple linear regression, 
#this coefficient is equal to the square of the correlation between the two variables.
with(iris.vir, cor(Sepal.Length,Petal.Length)^2)
```

```r
par(mfrow=c(2,2)) 
plot(modelA)
par(mfrow=c(1,1))
```

```r
shapiro.test(rstandard(modelA))
```

```r
#The assumption of uniform variance is not so clear from the plots, 
#particularly from the Scale-Location graph.
ncvTest(modelA)
```

```r
 ##there is no evidence for a non-constant variance of the residuals (heteroscedasticity)
```

```r
birds$logMaxAbund <- log10(birds$MaxAbund) 
birds$logMass <- log10(birds$Mass)
attach(birds)
par(mfrow=c(2,2))
hist(logMaxAbund, col = 'lightblue1',
xlab = 'MaxAbund', main = 'MaxAbund')
hist(logMass, col = 'lightblue1', xlab = 'Mass', main = 'Mass')
par(mfrow=c(1,1))
```

```r
shapiro.test(logMaxAbund)
```

```r
model2 <- lm(logMaxAbund ~ logMass, data = birds)
summary(model2)
```

```r
shapiro.test(rstandard(model2))
```

```r
plot(logMaxAbund ~ logMass, data = birds) 
abline(model2)
```

```r
par(mfrow=c(2,2))
plot(model2)
par(mfrow=c(1,1))
```

```r
#Subset the terrestrial birds and do a regression for them.
 model3 <- lm(logMaxAbund ~ logMass, data=birds, subset=birds$Aquatic == 0)
summary(model3)
```

```r
par(mfrow=c(2,2)) 
plot(model3)
par(mfrow=c(1,1)) 
```

```r
plot(logMaxAbund ~ logMass, data=birds, subset=!birds$Aquatic,
main="Terrestrial birds",
ylab = expression("log"[10]*"(Maximum Abundance)"), xlab = expression("log"[10]*"(Mass)"),
pch=19, col=rgb(176,196,222,200,maxColorValue=225))
abline(model3,lwd=2)

birds_t <- subset(birds,birds$Aquatic == 0) 
# yet another way of subsetting 
confit_model3<-predict(model3,interval="confidence")
I <- order(birds_t$logMass)
fit <- confit_model3[,1]
plusCI<-I(confit_model3[,3])
minusCI<-I(confit_model3[,2])
xx_model3 <- c(birds_t$logMass[I],rev(birds_t$logMass[I]))
yy_model3 <- c(plusCI[I],rev(minusCI[I]))
polygon(xx_model3,yy_model3,col=rgb(0,30,70,20,maxColorValue=225),border=rgb(0,30,70,20,maxColorValue=225))
```
