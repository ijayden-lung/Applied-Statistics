```r
library(car)
scatterplot(dist ~ speed, data = cars)
```

```r
model1 <- lm(dist ~ speed, data = cars)
summary(model1)
```

```r
(invXtX <- summary(model1)$cov.unscaled)
```

```r
##std
summary(model1)$sigma
```

```r
##variance
summary(model1)$sigma^2
```

```r
#covariance matrix
vcov(model1)
```

```r
##or
(summary(model1)$sigma^2)*invXtX
```

```r
confint(model1, level = 0.98)
```

```r
plot(dist ~ speed, data = cars)
abline(model1)
new.speed <- data.frame(speed=seq(3,27,
length.out = 15))
pc <- predict(model1,new.speed, int='c')
matlines(new.speed$speed, pc, lty=c(1,2,2),lwd=rep(2,3), col=c('black','red','red'))
pp <- predict(model1,new.speed, int='p') 
matlines(new.speed$speed, pp, lty=c(1,2,2),lwd=rep(2,3),
col=c('black','red','red'))
```
